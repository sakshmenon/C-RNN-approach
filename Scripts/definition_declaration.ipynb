```python
import re
def find_undefined_function_types(code_list):

    defined_types = set(['int', 'char', 'float', 'double', 'long', 'short', 'void'])
    undefined_types = set()
    defined_func_types = set()
    
    # Regular expression pattern to match function declarations
    function_pattern = r'\b([a-zA-Z_]\w*\s*\**)\s+\w+\s*\([^)]*\)\s*{'
    
    temp_list = [i for i in code_list]
    lines = "\n".join(temp_list)
    temp = lines
        # Find all function declarations in the content
    function_matches = re.findall(function_pattern, lines)
    # modified_lines = [lines]
    
    for return_type in function_matches:
        # Check if the return type is not in the set of defined types
        if return_type.strip('*') not in defined_types:
            undefined_types.add(return_type.strip('*'))

            temp = (temp.replace(return_type, 'int'))
        else:
            undefined_types.add(return_type.strip('*'))


    
    return list(undefined_types), list(defined_func_types), temp

def find_undefined_types_in_variables(code_list):

    cpp_keywords = [
    "alignas",
    "alignof",
    "and",
    "and_eq",
    "asm",
    "auto",
    "bitand",
    "bitor",
    "bool",
    "break",
    "case",
    "catch",
    "char",
    "char16_t",
    "char32_t",
    "class",
    "compl",
    "const",
    "constexpr",
    "const_cast",
    "continue",
    "decltype",
    "default",
    "delete",
    "do",
    "double",
    "dynamic_cast",
    "else",
    "enum",
    "explicit",
    "export",
    "extern",
    "false",
    "float",
    "for",
    "friend",
    "goto",
    "if",
    "inline",
    "int",
    "long",
    "mutable",
    "namespace",
    "new",
    "noexcept",
    "not",
    "not_eq",
    "nullptr",
    "operator",
    "or",
    "or_eq",
    "private",
    "protected",
    "public",
    "register",
    "reinterpret_cast",
    "return",
    "short",
    "signed",
    "sizeof",
    "static",
    "static_assert",
    "static_cast",
    "struct",
    "switch",
    "template",
    "this",
    "thread_local",
    "throw",
    "true",
    "try",
    "typedef",
    "typeid",
    "typename",
    "union",
    "unsigned",
    "using",
    "virtual",
    "void",
    "volatile",
    "wchar_t",
    "while",
    "xor",
    "xor_eq",
    "%d",
    "%f",
    "%s"
]

    defined_types = {'int', 'char', 'float', 'double', 'long', 'short', 'void'}
    undefined_types = set()
    defined_set = set()

    variable_pattern = r'\b((?:[a-zA-Z_]\w*\**)\s+\**\s*\**[a-zA-Z_]\w*\[*\w*\]*)\s*(?:,|\s*;|\s*=|\s*\))'

    for line in enumerate(code_list):
        # content = file.read()
        # Find all variable declarations in the content
        variable_matches = re.findall(variable_pattern, line[1])
        for variable_declaration in variable_matches:
            # Extract the data type from the variable declaration
            data_type = variable_declaration.split()[0]
            if data_type == 'd':
                # print(data_type)
                pass
            # Check if the data type is not in the set of defined types
            elif (data_type not in defined_types) and (data_type not in cpp_keywords):
                # print(variable_declaration.split()[0].replace(variable_declaration.split()[0], "int ") + variable_declaration.split()[1])

                undefined_types.add(data_type)
                code_list[line[0]][1] = (line[1][1].replace((variable_declaration.split()[0]+ ' '), 'int '))
            else:
                defined_set.add(data_type)



    return list(undefined_types), list(defined_set), code_list

def find_definition_statements(code):
    patterns = r'\w+\s*=\s*\d+'
    definition = []
    for line in code:
        matches = re.findall(patterns, line)
        if matches:
            definition.append({matches[0].split('=')[0]:matches[0].split('=')[1]})
    return definition

def find_declaration_statements(code):
    variable_pattern = r'\b((?:[a-zA-Z_]\w*\**)\s+\**\s*\**[a-zA-Z_]\w*\[*\w*\]*)\s*(?:,|\s*;|\s*=|\s*\))'
    key_list = ['int', 'bool', 'char']
    declarations = []
    for line in code:
        statements = re.findall(variable_pattern, line)
        if statements:
            if statements[0].split()[0] in key_list:
                declarations.append([statements[0].split()[0],statements[0].split()[1]])
    return declarations
```
